```
import os
from PIL import Image, ImageFilter
import numpy
```

```
# Criar uma pasta Mascaras que contém pastas com as classes. (Imagens com transparencia)
# Criar uma pasta background com as imagens que serão plano de fundo
dirForeground = "./Mascaras"
dirBackground = "./background"

def retornaListaBackgrounds(backgroundDir):
    nomesBackground = [f for f in os.listdir(backgroundDir) if os.path.isfile(os.path.join(backgroundDir, f))]
    return nomesBackground
	
listaBackgrounds = retornaListaBackgrounds(dirBackground)

# Proporção de imagens para cada classe em relação ao total de backgrounds
proporcao =[[0.25, 0.25, 0.25, 0.25],
            [0.25, 0.25, 0.25, 0.25]]
			
qntDeImagensPorClasse = int(len(listaBackgrounds)/ len(proporcao))
vetorRange = []

for linha in proporcao:
    arrayAux = []
    for coluna in linha:
        arrayAux.append(int(qntDeImagensPorClasse * coluna))
    vetorRange.append(arrayAux)

print(vetorRange)

l = 0
for root, dirs, files in os.walk(dirForeground):
    if root == dirForeground:
        continue
    diretorioFinal = './img' 
    if not os.path.exists(diretorioFinal):
        os.mkdir(diretorioFinal)
    c = 0
    for img in files:
        print(root+'/'+img)
        foreground = Image.open(root+'/'+img)
        print(str(l)+" "+str(c))
        for i in range(vetorRange[l][c]):
            foreground = foreground.rotate(numpy.random.randint(0, 75))
            background = Image.open(dirBackground+'/'+listaBackgrounds[i])
            backgroundResized = background.resize(foreground.size)
            # É possivel adicionar técnicas de data augmentation aqui
            if numpy.random.randint(0, 100) <= 15:
                # Adiciona Gaussian Blur
                backgroundResized = backgroundResized.filter(ImageFilter.GaussianBlur(radius=numpy.random.randint(10, 20)))

            backgroundResized.paste(foreground, (0, 0), foreground)
            imageComposed = backgroundResized.resize((224,224))
            imageComposed.save(diretorioFinal+'/'+str(root.split('/')[-1:][0])+'_'+ str(c)+'_'+str(i)+'.jpg')
            print(diretorioFinal+'/'+str(root.split('/')[-1:][0])+'_' + str(c)+'_'+str(i))
        del listaBackgrounds[0:vetorRange[l][c]]
        c += 1
l += 1
```
